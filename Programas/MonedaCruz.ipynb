```python
import random
def toss():
    return random.choice(["Heads", "Tails"])
```

```python
def exp(n):
    listContadores = list()
    for i in range(1,n):
        anterior = ""
        actual = ""
        contador = 0 
        actual = toss()
        contador=contador+1
        while True:
            anterior = actual
            actual = toss()
            contador=contador+1
            if actual=="Heads" and anterior=="Heads":
                break
        listContadores.append(contador)
    return sum(listContadores)/len(listContadores)
```

```python
exp(1000)
```

```python
dataX = [i for i in range(2,1000)]
dataY = [exp(i) for i in dataX]
```

```python
def yn(n):
    y=[]
    for k in range(n):
        nh=0
        c=0
        while True:
            r = int(np.random.random()+.5)
            c = c+1
            if r==1 and nh==1:
                break
            if r==1 and nh==0:
                nh=1
            if r==0:
                nh=0
        y.append(c)
    return sum(y)/n
```

```python
yn(1000)
```
