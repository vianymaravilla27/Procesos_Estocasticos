```python
# Importamos las librerias necesarias para correr el programa

import random
import matplotlib.pyplot as plt
```

```python
#  Proceso del randomwalk

def randomwalk(x,time):
    Ts = [0]; X = [0]
    for ts in range (1, time + 1):
        Ts.append(ts)
        rand = random.choice(x)
        step = X[ts-1] + rand
        X.append(step)
        
# Se procede a plotear lo obtenido en el randomwalk

    plt.figure(figsize = (8,6))
    plt.plot(Ts,X,label="Random Walk")
    plt.xlabel("Time")
    plt.ylabel("Walk")
    plt.show()

```

```python
randomwalk([-1,1],500)
```
