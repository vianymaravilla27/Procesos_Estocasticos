```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_mean_throws(n):
    results = []
    for _ in range(n):
        nh = 0
        throw_count = 0
        while True:
            r = np.random.randint(2)
            throw_count += 1
            if r == 1 and nh == 1:
                break
            elif r == 1 and nh == 0:
                nh = 1
            elif r == 0:
                nh = 0
        results.append(throw_count)
    return results

n = 10000
results = calculate_mean_throws(n)

# Calcular la media acumulada
mean_values = np.cumsum(results) / np.arange(1, n+1)

# Crear el gráfico
plt.plot(mean_values)
plt.axhline(y=6, color='r', linestyle='-')  # Dibujar una línea horizontal en y = 6
plt.title('Media acumulada de lanzamientos para obtener dos caras consecutivas')
plt.xlabel('Número de simulación')
plt.ylabel('Media de lanzamientos')
plt.show()
```
